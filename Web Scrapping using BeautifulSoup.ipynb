```python
from bs4 import BeautifulSoup
```

```python
import urllib
```

```python
link_init = "https://www.bankbazaar.com/reviews.html?reviewPageNumber="
```

```python
url = link_init + str(1)
headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/103.0.5060.114 Safari/537.36 Edg/103.0.1264.49'}
req = urllib.request.Request(url = url, headers = headers)
```

```python
doc = urllib.request.urlopen(req).read().decode('utf-8')
```

```python
doc[:100]
```

```python
bsobj = BeautifulSoup(doc)
```

```python
bsobj
```

```python
user_profile = container_0.find(name = 'div', attrs = {'class': 'reviewer-profile'}).text
```

```python
import re
```

```python
up = re.sub('Reviewed on ','', user_profile)
up
```

```python
up = re.sub('\\n', '', up)
up
```

```python
name = re.findall("^[A-Z ]+", up)[0].strip()
name
```

```python
date = re.findall("[A-Za-z]{3} [0-9]{2}, [0-9]{4}", up)[0]
date
```

```python
up1 = re.sub("Mar 21, 2020", "", up)
up1
```
